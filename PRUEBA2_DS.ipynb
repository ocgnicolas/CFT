```python
# Abrir la base de datos con la librería Pandas

import pandas as pd
data1 = pd.read_excel("prueba.xlsx")
```

```python
# Muestre las 10 primeras filas del dataframe

data1.head(10)
```

```python
# Muestre la longitud del dataframe

len(data1)
```

```python
data1s = data1
```

```python
# Renombre la columna "Beneficiarios/as Estimados a dic 2020\n 3,10 Benef/arr" por "beneficiarios"

data2 = data1s.rename(columns={"Beneficiarios/as Estimados a dic 2020\n 3,10 Benef/arr":"beneficiarios"})
```

```python
# Impute valores NaN de la columna anteriormente renombrada, mediante el promedio.

data2["beneficiarios"] = pd.to_numeric(data2["beneficiarios"])
data2["beneficiarios"].fillna(data2["beneficiarios"].mean())
```

```python
# Realice un subconjunto de datos con las columnas: "Comuna", "beneficiarios", "Tipo" y "Fondos" ("TIPO" MALO)

data2s = data2[["Comuna","beneficiarios","Fondos"]]
data2s.head()
```

```python
# Determine todas las columnas con mas de 3000 beneficiarios

cond = data2s[(data2s["beneficiarios"]>3000)]
cond.head()
```
